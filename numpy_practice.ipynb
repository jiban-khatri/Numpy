```python
# Create two arrays: a = [1, 2, 3, 4] and b = [10, 20, 30, 40]. Compute their element-wise sum, difference, and product.

a = np.array([1,2,3,4])
b = np.array([10,20,30,40])

print(a+b)
print(a-b)
print(a*b)
```

```python
# Create a 3×3 matrix. Compute its determinant, inverse, and solve the linear system Ax = b where b = [1, 2, 3].

A = np.random.randint(1,30,9).reshape(3,3)
A
```

```python
# determinant
det = np.round(np.linalg.det(A))
det
```

```python
# inverse
print(np.linalg.inv(A))
```

```python
#  AX = B
B = np.array([1,2,3])

X = (1/det) * (A.T * B)

print(X)
```
