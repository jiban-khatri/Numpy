```python
# CONCATENATE

c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d), axis = 0) #down wise
```

```python
np. concatenate((c,d), axis = 1)  #sidewise
```

```python
#  np.where -> returns index position acc to condition given

a
```

```python
# value > 50

np.where(a>50)
```

```python
#  replace grater than 50 with 0
# ----> np.where(condition, true, false)

np.where(a>50, 0, a)
```

```python
#  replaces all even numbers with 1
np.where(a%2==0, 1, a)
```
